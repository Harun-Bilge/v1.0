```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("C:/Users/harun/Desktop/project/task2/data/processed/enriched_dataset.csv")
```

```python
# Visualize how many sessions users typically complete on their first day
sns.histplot(df.drop_duplicates('user_id')['total_session_count'], bins=30)
plt.title("First-Day Session Count Distribution")
plt.show()
```

```python
# Track how session duration changes over time for each engagement segment
avg_duration = df.groupby(['event_date', 'engagement_segment'])['total_session_duration'].mean().reset_index()

sns.lineplot(data=avg_duration, x='event_date', y='total_session_duration', hue='engagement_segment')
plt.title("Avg Session Duration Over Time by Segment")
plt.show()
```

```python
# Compare average revenue across engagement segments
seg_revenue = df.groupby('engagement_segment')[['iap_revenue', 'ad_revenue']].mean().reset_index()

seg_revenue.plot(kind='bar', x='engagement_segment', figsize=(6,4))
plt.title("Average Revenue by Segment")
plt.show()
```

```python
# Show which countries generate the most revenue using a heatmap
country_rev = df.groupby('country')[['iap_revenue', 'ad_revenue']].sum()

sns.heatmap(country_rev, annot=False)
plt.title("Country vs Revenue Heatmap")
plt.show()
```
