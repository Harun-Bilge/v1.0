```python
import pandas as pd

df = pd.read_csv("C:/Users/harun/Desktop/project/task2/data/processed/cleaned_dataset.csv")
```

```python
# Pulling first day metrics
first_day = df.groupby('user_id').first().reset_index()

first_day = first_day[['user_id',
                       'total_session_count',
                       'total_session_duration',
                       'match_start_count',
                       'match_end_count',
                       'victory_count',
                       'defeat_count']]
```

```python
# Define a simple rule-based segmentation function for session count
def segment_user(x):
    if x <= 3:
        return "low"
    elif x <= 8:
        return "mid"
    else:
        return "high"
```

```python
# Assign an engagement segment to every user based on their session count
first_day['engagement_segment'] = first_day['total_session_count'].apply(segment_user)
```

```python
# Merge with full dataset
df = df.merge(first_day[['user_id', 'engagement_segment']], on='user_id', how='left')
```

```python
df.to_csv("C:/Users/harun/Desktop/project/task2/data/processed/enriched_dataset.csv", index=False)
```
