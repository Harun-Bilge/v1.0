```python
import pandas as pd
import numpy as np
import glob

# Load all CSV files from folder and merge into one dataframe

file_list = glob.glob("C:/Users/harun/Desktop/project/task2/data/raw/*.csv")

df_list = []

for file in file_list:
    temp = pd.read_csv(file)
    df_list.append(temp)

df = pd.concat(df_list, ignore_index=True)

df.head()
```

```python
# Print dataset structure and summary statistics to understand basic data quality
df.info()
df.describe()
```

```python
# Missing values check
df.isna().sum()
```

```python
# Replace missing revenue values with zero to avoid breaking numeric calculations
df['iap_revenue'] = df['iap_revenue'].fillna(0)
df['ad_revenue'] = df['ad_revenue'].fillna(0)
```

```python
# Convert date columns to datetime format and sorts events chronologically for each user
df['event_date'] = pd.to_datetime(df['event_date'])
df['install_date'] = pd.to_datetime(df['install_date'])
df = df.sort_values(by=['user_id', 'event_date'])
```

```python
df.to_csv("C:/Users/harun/Desktop/project/task2/data/processed/cleaned_dataset.csv", index=False)
```
